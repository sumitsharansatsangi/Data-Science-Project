# Page Rank Algorithm

### Importing libraries

```python
import networkx as nx
import matplotlib.pyplot as mat
import math
import random
import numpy as np
import pandas as pd
```

### Creating adjacency matrix  

```python
n=10
init_prob=[1/n]*n 
adj_matrix=[]
adj_list=[]
for i in range(n):
    adj1=[]
    for j in range(n):
        adjl=[]
        if(random.random()>0.5):
            #appending adjacency matrix
            adj1.append(1)
        else:
            adj1.append(0)    
    adj_matrix.append(adj1)
```

## Plotting The Graph Circularly

```python
adj=pd.DataFrame(adj_matrix)
G = nx.DiGraph(adj.values)
nx.draw_circular(G)
```

## Plotting The Graph normally

```python
nxG = nx.DiGraph(adj.values)
nx.draw(nxG)
```

## assigning Transition Probability

```python
trans_prob=[]
for i in range(n):
    trans_prob.append(list(np.array(adj_matrix[i])/sum(adj_matrix[i])))
pd.DataFrame(trans_prob)
```

## Calculating Page Rank 

```python
init_prob_prev=[]
zero_array=[0]*n
for i in range(1000):
    init_prob_prev=init_prob
    init_prob=np.dot(init_prob,trans_prob)
    if(list(init_prob-init_prob_prev)==zero_array):
        print(i)
        break
    
```

## Updated Probability

```python
for i in range(999):
    init_prob=np.dot(init_prob,trans_prob)
```
